```python
from graphviz import Graph

def plotHeap(T):
    dot = Graph(comment='Heap')
    dot.node_attr.update(fillcolor='white', style='filled')
    dot.node_attr.update(shape='circle', heigth = '0.6', width='0.6',fixedsize='true')
    dot.attr(bgcolor='transparent')

    for i,v in enumerate(T):
        dot.node(i.__str__(),v.__str__(),fontname='courier',fontsize='22')

    for i in range(0,len(T)//2):
        e1 = 2*i+1
        e2 = 2*i+2
        dot.edge(str(i),str(e1))
        if e2 < len(T):
            if e2 < len(T):
                dot.edge(str(i),str(e2))
            else:
                dot.edge(str(i),str(e2),color='white')
            
    dot.engine = 'dot'        
    dot.render('tmp/heap.gv', view=True)  # doctest: +SKIP 
    return dot
```

```python
T = [ 8, 5, 7, 4, 1, 6, 3, 2 ]; plotHeap(T)
```
